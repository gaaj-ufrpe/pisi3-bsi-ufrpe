#Introdução

O pandas é uma lib utilizada para trabalhar sobre dataframes, que são tabelas de dados.

```python
import traceback
import pandas as pd
import numpy as np
import random
import string
```

```python
random.seed(42)
def gen_id():
   chars = string.ascii_lowercase + string.digits
   return ''.join(random.choices(chars, k=10))

n = 50
df_pessoas = pd.DataFrame({
    'ID': [gen_id() for _ in range(n)],
    'Nome': [f'Pessoa {x}' for x in range(1,n+1)],
    'Idade': list(random.choices(range(15,30),k=n)),
    'gênero': list(random.choices(['F','M','O'],weights=[.5,.45,.05],k=n)),
    'raça ': list(random.choices(['Preta','Branca','Latina','Indígena','Amarela','Outra',None],
                                weights=[.2,.25,.2,.1,.1,.1,.05],
                                k=n)),
    'ESTADO CIVIL': list(random.choices(['Solteiro','Casado','Divorciado','Viuvo','Outro',None],k=n)),
    'altura': [np.round(random.gauss(1.7,.15), 2) for _ in range(n)],
    'peso': [np.round(random.gauss(70,10),1) for _ in range(n)],
})
print('A função "head" mostra os primeiros registros da tabela. A função "tail" mostra os últimos registros da tabela. A função "sample" retorna uma amostra aleatória.')
print()
print('df_pessoas.head()\n',df_pessoas.head())
print()
print('df_pessoas.head(10)\n',df_pessoas.head(10))
print()
print('df_pessoas.tail(3)\n',df_pessoas.tail(3))
print()
print('df_pessoas.sample(3)\n',df_pessoas.sample(3))
```

O método `rename` pode ser usado para renomear algumas colunas do dataframe. Este método, assim como outros, possui uma flag `inplace` para indicar se a alteração deve ser feita no dataframe original, ou se será retornado um novo dataframe.

# Incluindo Colunas

É possível operar facilmente sobre os dados do dataframe.
Incluindo colunas:

```python
df_pessoas['ativo'] = 1
df_pessoas['maior_idade'] = df_pessoas['idade']>=18
df_pessoas['imc'] = np.round(df_pessoas['peso']/(df_pessoas['altura']**2), 1)
display(df_pessoas.head())
```

Incluindo uma coluna `classe_imc` no dataframe, com base na seguinte tabela de classificação do IMC:
<table>
<tr>
  <th>Valor do IMC</th>
  <th>Classificação</th>
</tr>
<tr>
  <td>Menor que 18,5</td>

<tr>

  <td>Obesidade grau I</td>

```python
def classify_imc(imc):
  if imc < 18.5:
    result = 'Magreza'
  elif imc < 25:
    result = 'Normal'
  elif imc < 30:
    result = 'Sobrepeso'
  elif imc < 35:
    result = 'Obesidade grau I'
  elif imc < 40:
    result = 'Obesidade grau II'
  else:
    result = 'Obesidade grau III'
  return result

df_pessoas['classe_imc'] = df_pessoas['imc'].apply(classify_imc)
display(df_pessoas.head())
```

Também é possível usar expressões lambda junto com o `apply`.
O `apply` pode ser aplicado sobre a coluna, ou mesmo sobre o dataframe. Neste caso, o parâmetro `axis=1` faz com que cada elemento seja uma linha da tabela, enquanto `axis=0` faz com que cada elemento seja uma coluna.

```python
df_pessoas['obesidade'] = df_pessoas['classe_imc'].apply(lambda x: 'Obesidade' in x)
df_pessoas['obesidade_menor_idade'] = df_pessoas.apply(
    lambda row: 'Obesidade' in row['classe_imc'] and row['idade']<18,
    axis=1)
display(df_pessoas.head())
```

O método `map` é utilizado para mapear um valor a outro utilizando um dicionário. A chave é o valor original e o valor será o novo valor que será colocado no dataframe.

```python
df_pessoas['genero_desc'] = df_pessoas['genero'].map({
    'F':'Feminino','M':'Masculino','O':'Outro',
})
display(df_pessoas.head())
```

# Manipulando Valores Vazios

A partir de agora, vamos usar o método `copy` para não alterar o dataframe original.
No trecho abaixo, substituímos as células nulas por 'VAZIO'.

```python
df_aux = df_pessoas.copy()
print('Shape antes do fillna:',df_aux.shape)
print('Vazios antes do fillna:',len(df_aux[df_aux.isna().any(axis=1)]))
df_aux.fillna('VAZIO', inplace=True)
print('Shape depois do fillna:',df_aux.shape)
print('Vazios depois do fillna:',len(df_aux[df_aux.isna().any(axis=1)]))
```

# Group by

O método `groupby` pode ser usado para agrupar um dataframe a partir de uma ou mais colunas, agregando ass demais usando funções como `sum`, `count`, `mean`, `std`, `nunique`...

O uso do `copy` após selecionar o `subset` do dataframe, evita efeitos colaterais de mudança nos dados do dataframe original caso se manipule os dados do `subset`.

```python
cols = ['customer_type','total','cogs','gross_income']
df_group = df_sales[cols].copy()
df_group = df_group.groupby(by='customer_type').agg(['count','mean','std','sum'])
display(df_group)
```

Usando o `NamedAgg` é possível criar um nome para cada coluna agregada e uma função específica para cada. Com isso, fica mais fácil recuperar os valores das colunas pelo nome posteriormente.

```python
cols = ['customer_type','total','cogs','gross_income']
df_group = df_sales[cols].copy()
df_group = df_group.groupby(by='customer_type').agg(
    total_count=pd.NamedAgg('total','count'),
    total_mean=pd.NamedAgg('total','mean'),
    total_std=pd.NamedAgg('total','std'),
    total_sum=pd.NamedAgg('total','sum'),
    cogs_count=pd.NamedAgg('cogs','count'),
    cogs_mean=pd.NamedAgg('cogs','mean'),
    cogs_std=pd.NamedAgg('cogs','std'),
    cogs_sum=pd.NamedAgg('cogs','sum'),
    gross_income_count=pd.NamedAgg('gross_income','count'),
    gross_income_mean=pd.NamedAgg('gross_income','mean'),
    gross_income_std=pd.NamedAgg('gross_income','std'),
    gross_income_sum=pd.NamedAgg('gross_income','sum'),
)
display(df_group)
```

Ao agrupar com o `NamedAgg`, fica mais fácil para recuperar as colunas criadas pelo nome.

```python
print(df_group[['total_mean','total_std']])
```

Contudo, as colunas utilizadas no `groupby` não são colunas no dataframe resultante, mas índices do dataframe. Por isso que o código abaixo dá erro.

```python
try:
  print(df_group[['customer_type','total_mean','total_std']])
except Exception as e:
  traceback.print_exception(e)
```

Caso seja necessário operar sobre as colunas usadas no `groupby`, é interessante resetar os índices criados pelo `groupby`. Observe a diferença abaixo.

```python
df_group2 = df_group.reset_index()
display(df_group2)
print(df_group2[['customer_type','total_mean','total_std']])
```
